```python
import os
import glob
import pandas as pd

# Lire le CSV dans un DataFrame
df = pd.read_csv('votre_fichier.csv')

# Créer des listes pour stocker les valeurs des nouvelles colonnes
noms_recent = []
dates_creation = []

# Parcourir chaque ligne du DataFrame
for index, row in df.iterrows():
    chemin_dossier = row['chemin']
    nom_recherche = row['nom']

    # Construire le chemin absolu du dossier
    chemin_absolu = os.path.abspath(chemin_dossier)

    # Utiliser la bibliothèque glob pour trouver tous les fichiers correspondants
    fichiers_correspondants = glob.glob(os.path.join(chemin_absolu, f'*{nom_recherche}*'))

    if fichiers_correspondants:
        # Trouver le fichier le plus récent parmi ceux correspondants
        fichier_plus_recent = max(fichiers_correspondants, key=os.path.getctime)
        nom_plus_recent = os.path.basename(fichier_plus_recent)
        date_creation = os.path.getctime(fichier_plus_recent)

        noms_recent.append(nom_plus_recent)
        dates_creation.append(date_creation)
    else:
        noms_recent.append(None)
        dates_creation.append(None)

# Ajouter les nouvelles colonnes au DataFrame
df['nom_recent'] = noms_recent
df['date_creation'] = dates_creation

# Enregistrer le DataFrame mis à jour dans un nouveau fichier CSV
df.to_csv('nouveau_fichier.csv', index=False)
```

```python
import pandas as pd
import re

# Exemple de données
data = {'colonne_objet': ['abc123456789def', 'x45y67z', '12abc34', 'abc',None]}

df = pd.DataFrame(data)

# Fonction pour extraire les 6 premiers chiffres
def extract_first_six_digits(s):
    if s:
        digits = re.findall(r'\d', s)
        if digits:
            num_str = ''.join(digits[:6])
            return num_str
        else:
            return None

# Appliquer la fonction à la colonne 'colonne_objet'
df['chiffres_extraits'] = df['colonne_objet'].apply(extract_first_six_digits)

# Afficher le DataFrame mis à jour
print(df)
```

```python
import pandas as pd
import numpy as np

# Exemple de données
data = {'colonne_objet': ['abc123456def', 'x45y67z', '12abc34', 'nan', np.nan]}

df = pd.DataFrame(data)

# Fonction pour extraire les 6 premiers chiffres
def extract_first_six_digits(s):
    if isinstance(s, str) and s.lower() != 'nan':  # Vérifie si la valeur est une chaîne non 'nan'
        digits = [c for c in s if c.isdigit()]
        if digits:
            num_str = ''.join(digits[:6])
            return int(num_str)
    return None

# Appliquer la fonction à la colonne 'colonne_objet'
df['chiffres_extraits'] = df['colonne_objet'].apply(extract_first_six_digits)

# Afficher le DataFrame mis à jour
print(df)
```

```python
import os
import pandas as pd

def find_recent_files(df):
    recent_files = []

    for index, row in df.iterrows():
        folder_path, filename_approx = row['le_chemin_du_dossier'], row['le_nom_aproximatif']
        matching_files = [f for f in os.listdir(folder_path) if filename_approx in f]
        
        if matching_files:
            matching_files.sort(key=lambda x: os.path.getmtime(os.path.join(folder_path, x)), reverse=True)
            recent_files.extend(matching_files[:2])  # Adding the two most recent files
    
    return recent_files

# Example DataFrame
data = {'le_chemin_du_dossier': ['/path/to/folder1', '/path/to/folder2'],
        'le_nom_aproximatif': ['example', 'sample']}
df = pd.DataFrame(data)

recent_files = find_recent_files(df)
print(recent_files)
```

```python
import os
from datetime import datetime

def trouver_fichiers_recentes(chemin_dossier, nom_racine):
    fichiers_trouves = []
    
    # Parcours du dossier
    for root, dirs, files in os.walk(chemin_dossier):
        for fichier in files:
            if fichier.startswith(nom_racine):
                chemin_fichier = os.path.join(root, fichier)
                date_creation = datetime.fromtimestamp(os.path.getctime(chemin_fichier))
                fichiers_trouves.append((fichier, date_creation, chemin_fichier))
    
    # Tri des fichiers par date de création
    fichiers_trouves.sort(key=lambda x: x[1], reverse=True)
    
    # Récupération des deux fichiers les plus récents
    fichiers_recents = fichiers_trouves[:2]
    
    return fichiers_recents

# Exemple d'utilisation
chemin_dossier = "/chemin/vers/votre/dossier"
nom_racine = "nom_racine"
resultat = trouver_fichiers_recentes(chemin_dossier, nom_racine)

for fichier, date_creation, chemin_fichier in resultat:
    print("Nom du fichier:", fichier)
    print("Date de création:", date_creation)
    print("Chemin du fichier:", chemin_fichier)
    print("=" * 40)
```

```python
import os
from datetime import datetime

def find_most_recent_file(chemin_dossier, nom_recherche):
    fichier_le_plus_recent = None
    date_plus_recente = None
    chemin_acces_input = None

    for fichier in os.listdir(chemin_dossier):
        if nom_recherche in fichier:
            chemin_fichier = os.path.join(chemin_dossier, fichier)
            date_creation = datetime.fromtimestamp(os.path.getctime(chemin_fichier))

            if date_plus_recente is None or date_creation > date_plus_recente:
                date_plus_recente = date_creation
                fichier_le_plus_recent = fichier
                chemin_acces_input = chemin_fichier

    return fichier_le_plus_recent, date_plus_recente, chemin_acces_input

# Example usage
chemin_dossier = r"your_directory_path_here"
nom_recherche = "search_criteria_here"
fichier_le_plus_recent, date_plus_recente, chemin_acces_input = find_most_recent_file(chemin_dossier, nom_recherche)

print("Most recent file:", fichier_le_plus_recent)
print("Date of creation:", date_plus_recente)
print("Path to file:", chemin_acces_input)
```

```python
import os
from datetime import datetime

def get_last_file_name(chemin_dossier, nom_recherche):
    fichier_le_plus_recent = None
    ficher_2_le_plus_recent = None
    date_plus_recente = None
    date_2_plus_recente = None
    chemin_acces_input = None
    chemin_2_acces_input = None

    try:
        for fichier in os.listdir(chemin_dossier):
            if nom_recherche in fichier:
                chemin_fichier = os.path.join(chemin_dossier, fichier)
                date_creation = datetime.fromtimestamp(os.path.getctime(chemin_fichier))

                if date_2_plus_recente is None or date_creation > date_2_plus_recente:
                    date_2_plus_recente = date_creation
                    ficher_2_le_plus_recent = fichier
                    chemin_2_acces_input = chemin_fichier

                    if date_plus_recente is None or date_2_plus_recente > date_plus_recente:
                        date_plus_recente = date_2_plus_recente
                        fichier_le_plus_recent = ficher_2_le_plus_recent
                        chemin_acces_input = chemin_2_acces_input

        return fichier_le_plus_recent, date_plus_recente, chemin_acces_input, ficher_2_le_plus_recent, date_2_plus_recente

    except OSError as e:
        print("Error:", e)
        return None

# Example usage
chemin_dossier = r"your_directory_path_here"
nom_recherche = "search_criteria_here"
result = get_last_file_name(chemin_dossier, nom_recherche)

if result is not None:
    fichier_le_plus_recent, date_plus_recente, chemin_acces_input, ficher_2_le_plus_recent, date_2_plus_recente = result
    print("Most recent file:", fichier_le_plus_recent)
    print("Date of creation:", date_plus_recente)
    print("Path to file:", chemin_acces_input)
```

```python
import os
from datetime import datetime

def get_last_file_info(directory_path, root_name, num_files=2):
    try:
        file_info_list = []

        for root, dirs, files in os.walk(directory_path):
            for file in files:
                if file.startswith(root_name):
                    file_path = os.path.join(root, file)
                    creation_date = datetime.fromtimestamp(os.path.getctime(file_path))
                    file_info_list.append((file, creation_date, file_path))

        file_info_list.sort(key=lambda x: x[1], reverse=True)
        recent_files = file_info_list[:num_files]

        return recent_files

    except Exception as e:
        print("An error occurred:", e)
        return []

# Example usage
directory = "/path/to/your/directory"
root = "desired_root_name"
recent_files_info = get_last_file_info(directory, root)

for file_info in recent_files_info:
    file_name, creation_date, file_path = file_info
    print("File Name:", file_name)
    print("Creation Date:", creation_date)
    print("File Path:", file_path)
    print("----------")
```
