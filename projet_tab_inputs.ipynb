```python
import os
import pandas as pd
from datetime import datetime, timedelta
```

```python
repertoire = 'repertoire_xlsx'
```

```python
for i in range(1, 51):
    nom_fichier = f'fiche_{i}.xlsx'
    chemin_fichier = os.path.join(repertoire, nom_fichier)
    
    # Création d'un DataFrame vide
    data = {'Colonne1': [], 'Colonne2': [], 'Colonne3': []}
    df = pd.DataFrame(data)
    
    # Enregistrement du DataFrame dans un fichier Excel
    df.to_excel(chemin_fichier, index=False)
```

```python
data = []

for nom_fichier in os.listdir(repertoire):
    chemin_fichier = os.path.join(repertoire, nom_fichier)
    if os.path.isfile(chemin_fichier):
        date_creation = datetime.fromtimestamp(os.path.getctime(chemin_fichier))
        date_aujourd_hui = datetime.now()
        ecart_dates = date_aujourd_hui - date_creation
        data.append([nom_fichier, date_creation, date_aujourd_hui, ecart_dates])
```

```python
df = pd.DataFrame(data, columns=['Nom du fichier', 'Date de création', 'Date d\'aujourd\'hui', 'Écart de dates'])
```
